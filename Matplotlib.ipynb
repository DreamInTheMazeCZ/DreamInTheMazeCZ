```python
plt.plot(np.random.randn(50).cumsum()); # cumsum() : 누계
```

● 라인 스타일(Line Style)
  - 문자      스타일
  - '-'   :   'solid'
  - '--'  :   'dashed'
  - '-.'  :   'dashdot'
  - ':'   :   'dotted'

```python
# 단어 입력
plt.plot(np.random.randn(50).cumsum(), linestyle = 'solid')    # Blue
plt.plot(np.random.randn(50).cumsum(), linestyle = 'dashed')   # Orange
plt.plot(np.random.randn(50).cumsum(), linestyle = 'dashdot')  # Green
plt.plot(np.random.randn(50).cumsum(), linestyle = 'dotted');  # Red
```

```python
# 기호 입력
plt.plot(np.random.randn(50).cumsum(), linestyle = '-')   # Blue
plt.plot(np.random.randn(50).cumsum(), linestyle = '--')  # Orange
plt.plot(np.random.randn(50).cumsum(), linestyle = '-.')  # Green
plt.plot(np.random.randn(50).cumsum(), linestyle = ':');  # Red
```

● 색상 스타일(Color Style)
   - 문자      색상
   - 'b'    :  blue
   - 'g'    :  green
   - 'r'    :  red
   - 'c'    :  cyan
   - 'm'    :  magenta
   - 'y'    :  yellow
   - 'k'    :  black
   - 'w'    :  white

```python
# 다양한 입력 가능 : HTML 색상표, 실수 표현, 색 명칭 등
plt.plot(np.random.randn(50).cumsum(), color = 'g')
plt.plot(np.random.randn(50).cumsum(), color = '#1243FF')
plt.plot(np.random.randn(50).cumsum(), color = (0.2, 0.4, 0.6))
plt.plot(np.random.randn(50).cumsum(), color = 'darkblue');
```

```python
plt.plot(np.random.randn(50).cumsum(), color = 'skyblue')
plt.plot(np.random.randn(50).cumsum(), color = 'dodgerblue')
plt.plot(np.random.randn(50).cumsum(), color = 'royalblue')
plt.plot(np.random.randn(50).cumsum(), color = 'navy');
```

```python
# 라인 스타일, 컬러 스타일 기호로 동시 표현 가능
plt.plot(np.random.randn(50).cumsum(), 'b-')
plt.plot(np.random.randn(50).cumsum(), 'g--')
plt.plot(np.random.randn(50).cumsum(), 'c-.')
plt.plot(np.random.randn(50).cumsum(), 'm:');
```

● 플롯 축(Plot Axis)

```python
plt.plot(np.random.randn(50));
```

```python
plt.plot(np.random.randn(50));
plt.xlim(-1, 50)   # X축 범위 지정
plt.ylim(-5, 5);   # Y축 범위 지정
```

```python
# .axis를 통해 list로 지정 가능
plt.plot(np.random.randn(50));
plt.axis([-1, 50, -5, 5]);
```

```python
plt.plot(np.random.randn(50));
plt.axis('tight');   # 축에 가깝게 표현
```

```python
plt.plot(np.random.randn(50));
plt.axis('equal');   # 축에 멀게 표현
```

● 플롯 레이블(Plot Label)

```python
plt.plot(np.random.randn(50))
plt.title("Title")          # 차트의 이름
plt.xlabel("X")             # X축 이름
plt.ylabel("random.randn"); # Y축 이름
```

```python
plt.plot(np.random.randn(50), label = 'A') # 범례 설정
plt.plot(np.random.randn(50), label = 'B')
plt.plot(np.random.randn(50), label = 'C')
plt.title("Title")         # 차트의 이름
plt.xlabel("X")            # X축 이름
plt.ylabel("random.randn") # Y축 이름
plt.legend();              # 범례
```
