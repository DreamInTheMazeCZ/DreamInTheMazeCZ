# One-Hot-Encoding

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

data = pd.read_csv('../BigData_data/vote.csv', encoding='utf-8')
data.head()

# 범주형 변수 gender / region이 One-Hot-Encoding 대상
```

```python
X1 = data[data.columns[0:2]]
XY = data[data.columns[2:]]
```

## 범주형 변수 변환 - One-Hot-Encoding

```python
X1['gender'] = X1['gender'].replace([1,2], ['male', 'female'])
X1['region'] = X1['region'].replace([1, 2, 3, 4, 5], ['Sudo', 'Chungcheung', 'Honam', 'Youngnam', 'Others'])
```

```python
X1.head()
```

```python
# get_dummies()를 이용한 One-Hot-Encoding
X1_dum = pd.get_dummies(X1)
X1_dum.head()
```

## 자료 통합 및 저장

```python
Fvote=pd.concat([X1_dum, XY], axis=1)
Fvote.head()
```

```python
Fvote.to_csv('Fvote.csv', index=False, sep=',', encoding='utf-8')
```
